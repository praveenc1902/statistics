Central Tendency 
----------------
Is the central value or typical value in the dataset, It is the value that represents the data
there are 3 types
MEAN, MEDIAN, MODE . They are mostly used in missing values 

Mean - Sum of all value / total no of values 

Median - Order the dataset and pick the middle value, if there are 2 middle values then, avg of those 2 middle values 

Mode - Is the value which occurs most ofen

Measure Of Variablity 
-----
Is to find out how the data is distributed in the dataset
We have 3 ways to find the variability in the data 
1. Range --> Difference b/w max value and min value 
2. Variance --> Is a measure of how much distance is the each value from the mean 
3. Std Deviation --> is same as variance , it is just the square root of dataset

Percentiles and Quantiles
-----
Percentile -->  How much percentage of values are below that percent, if we say 50 percentile then that means, 50 % of the values are below that percentile

Quantile  --->  how many values are present b/w the certain limit





Correlation & Causation
---
Correlation -- Extend to which the 2 variables are realted. But, it doesn't mean, once variable is cause of other

+ve correaltion -- one variable increases, other vraible decreases 

-ve correlation -- once variable increases, other variable decreases


Causaation
-----
Once event causes the other event to occur. Thus there is a relationship b/w cause and effect b/w 2 variables



```python
from sklearn.datasets import load_boston
import pandas as pd 
```

```python
boston = load_boston()
df = pd.DataFrame(boston.data, columns = boston.feature_names)
df.head()
```

```python
df['price'] = boston.target
df.head()
```

```python
df_corr = df.corr()
```

```python
df_corr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline 
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df_corr,annot=True)
```
